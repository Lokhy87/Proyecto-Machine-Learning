```python
def simular_batalla(features, model):
    """
    Función para simular una batalla entre dos héroes.

    Parameters:
    features (dict): Diccionario con las características de los héroes.
    model (sklearn model): Modelo de Random Forest cargado.

    Returns:
    str: Resultado de la batalla.
    """
    # Convertir características a DataFrame
    features_df = pd.DataFrame(features, index=[0])
    # Realizar predicción
    prediction = model.predict(features_df)
    # Interpretar resultado
    if prediction[0] == 1:
        return "Heroe 1 gana"
    else:
        return "Heroe 2 gana"

# Uso de la función con el nuevo combate
result = simular_batalla(nuevo_combate, model)
print(result)
```

```python
import random

def generar_combate_aleatorio():
    return {
        'Heroe_1_encoded': random.randint(1, 500),
        'Intelligence_1': random.randint(1, 100), 
        'Strength_1': random.randint(1, 100), 
        'Speed_1': random.randint(1, 100), 
        'Durability_1': random.randint(1, 100), 
        'Power_1': random.randint(1, 100),
        'Combat_1': random.randint(1, 100),
        'Heroe_2_encoded': random.randint(1, 500), 
        'Intelligence_2': random.randint(1, 100),
        'Strength_2': random.randint(1, 100),
        'Speed_2': random.randint(1, 100),
        'Durability_2': random.randint(1, 100), 
        'Power_2': random.randint(1, 100),
        'Combat_2': random.randint(1, 100)
    }

# Generar una lista de combates aleatorios
combates_aleatorios = [generar_combate_aleatorio() for _ in range(5)]

for i, combate in enumerate(combates_aleatorios):
    print(f"Combate {i+1}: {combate}")
```
