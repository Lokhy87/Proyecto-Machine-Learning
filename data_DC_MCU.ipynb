```python
# Verificar las dimensiones del dataframe original
original_entries = len(df_new_results)

# Verificar si hay duplicados en las columnas de héroes en los datos de atributos
unique_heroes_attributes = df_heroes['nombre'].nunique()

# Verificar cuántos héroes del nuevo_battle_results no tienen coincidencia en el data_mcu_dc
heroe_1_not_in_attributes = df_new_results[~df_new_results['Heroe 1'].isin(df_heroes['nombre'])].shape[0]
heroe_2_not_in_attributes = df_new_results[~df_new_results['Heroe 2'].isin(df_heroes['nombre'])].shape[0]

# Resultados
original_entries, unique_heroes_attributes, heroe_1_not_in_attributes, heroe_2_not_in_attributes
```

```python
# Select the relevant columns from data_mcu_dc
hero_attributes = df_heroes[['nombre', 'universo', 'Intelligence', 'Strength', 'Speed', 'Durability', 'Power', 'Combat', 'superpoderes']]

# Rename columns to avoid confusion during merge
hero_attributes.rename(columns={
    'nombre': 'Hero',
    'universo': 'Universe',
    'Intelligence': 'Intelligence',
    'Strength': 'Strength',
    'Speed': 'Speed',
    'Durability': 'Durability',
    'Power': 'Power',
    'Combat': 'Combat',
    'superpoderes': 'Superpowers'
}, inplace=True)

# Merge attributes for Heroe 1
nuevo_battle_results_merged = df_new_results.merge(hero_attributes, left_on='Heroe 1', right_on='Hero', how='left')

# Rename columns to specify Heroe 1 attributes
nuevo_battle_results_merged.rename(columns={
    'Universe': 'Heroe 1 Universe',
    'Intelligence': 'Heroe 1 Intelligence',
    'Strength': 'Heroe 1 Strength',
    'Speed': 'Heroe 1 Speed',
    'Durability': 'Heroe 1 Durability',
    'Power': 'Heroe 1 Power',
    'Combat': 'Heroe 1 Combat',
    'Superpowers': 'Heroe 1 Superpowers'
}, inplace=True)

# Drop the extra 'Hero' column
nuevo_battle_results_merged.drop(columns=['Hero'], inplace=True)

# Merge attributes for Heroe 2
nuevo_battle_results_merged = df_new_results.merge(hero_attributes, left_on='Heroe 2', right_on='Hero', how='left')

# Rename columns to specify Heroe 2 attributes
nuevo_battle_results_merged.rename(columns={
    'Universe': 'Heroe 2 Universe',
    'Intelligence': 'Heroe 2 Intelligence',
    'Strength': 'Heroe 2 Strength',
    'Speed': 'Heroe 2 Speed',
    'Durability': 'Heroe 2 Durability',
    'Power': 'Heroe 2 Power',
    'Combat': 'Heroe 2 Combat',
    'Superpowers': 'Heroe 2 Superpowers'
}, inplace=True)

# Drop the extra 'Hero' column
nuevo_battle_results_merged.drop(columns=['Hero'], inplace=True)

```
