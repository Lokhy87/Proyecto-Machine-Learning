DATA BATTLE_RESULTS

```python
df_results = pd.read_csv('./data/Battle_results.csv')
```

```python
df_results.info()
df_results.head()
```

```python
df_results.iloc[500:510]
```

```python
# Funcion visualizar enfrentamiento concreto
def visualizar_enfrentamiento(df_results, heroe1, heroe2):
    enfrentamiento = df_results[(df_results['Heroe 1'] == heroe1) & (df_results['Heroe 2'] == heroe2)]
    if enfrentamiento.empty:
        enfrentamiento = df_results[(df_results['Heroe 1'] == heroe2) & (df_results['Heroe 2'] == heroe1)]
    
    if not enfrentamiento.empty:
        print(enfrentamiento)
    else:
        print(f"No se encontr√≥ un enfrentamiento entre {heroe1} y {heroe2}")
```

```python
# Ejemplo
visualizar_enfrentamiento(df_results, 'YellowJacket', 'Thor')
```

Merge para crear data final

```python
df_heroes.info()
```

```python
df_results.info()
```

```python
# Asegurarse de que los nombres de los h√©roes sean √∫nicos en df_heroes
df_heroes = df_heroes.drop_duplicates(subset=['nombre'])

# Realizar la primera fusi√≥n
merge_data_heroes_1 = df_results.merge(df_heroes, left_on='Heroe 1', right_on='nombre', how='left', suffixes=('_1', '_2'))

# Realizar la segunda fusi√≥n
merge_data_final = merge_data_heroes_1.merge(df_heroes, left_on='Heroe 2', right_on='nombre', how='left', suffixes=('_1', '_2'))

# Seleccionar y renombrar columnas para evitar duplicaci√≥n y confusi√≥n
merge_data_final = merge_data_final[[
    'Heroe 1','universo_1', 'Intelligence_1', 'Strength_1', 'Speed_1', 'Durability_1', 'Power_1', 'Combat_1', 'Tier_1', 'Tier2_1', 'superpoderes_1',
    'Heroe 2', 'universo_2', 'Intelligence_2', 'Strength_2', 'Speed_2', 'Durability_2', 'Power_2', 'Combat_2', 'Tier_2', 'Tier2_2', 'superpoderes_2',
    'Ganador'
]]
```

```python
merge_data_final.info()
merge_data_final.head()
```

Data final para trabajar

```python
merge_data_final.to_csv('./data/Data_final_MCU_vs_DC.csv', index = False)
```
