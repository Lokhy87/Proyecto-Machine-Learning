```python
# Funcion visualizar enfrentamiento concreto
def visualizar_enfrentamiento(df_new_results, heroe1, heroe2):
    enfrentamiento = df_new_results[(df_new_results['Heroe 1'] == heroe1) & (df_new_results['Heroe 2'] == heroe2)]
    if enfrentamiento.empty:
        enfrentamiento = df_new_results[(df_new_results['Heroe 1'] == heroe2) & (df_new_results['Heroe 2'] == heroe1)]
    
    if not enfrentamiento.empty:
        print(enfrentamiento)
    else:
        print(f"No se encontró un enfrentamiento entre {heroe1} y {heroe2}")
```

```python
# Ejemplo
visualizar_enfrentamiento(df_new_results, 'YellowJacket', 'Thor')
```

```python
# Verificar las dimensiones del dataframe original
original_entries = len(df_new_results)

# Verificar si hay duplicados en las columnas de héroes en los datos de atributos
unique_heroes_attributes = df_heroes['nombre'].nunique()

# Verificar cuántos héroes del nuevo_battle_results no tienen coincidencia en el data_mcu_dc
heroe_1_not_in_attributes = df_new_results[~df_new_results['Heroe 1'].isin(df_heroes['nombre'])].shape[0]
heroe_2_not_in_attributes = df_new_results[~df_new_results['Heroe 2'].isin(df_heroes['nombre'])].shape[0]

# Resultados
original_entries, unique_heroes_attributes, heroe_1_not_in_attributes, heroe_2_not_in_attributes
```

```python
# Identificar héroes duplicados en el dataframe original data_mcu_dc
duplicate_heroes = df_heroes[df_heroes.duplicated(subset=['nombre'], keep=False)].sort_values(by='nombre')

# Mostrar los héroes duplicados
duplicate_heroes[['nombre_completo', 'nombre', 'universo']]
```

```python

# Verificar si hay duplicados exactos en el DataFrame completo
exact_duplicates = df_heroes[df_heroes.duplicated()]

# Verificar si hay duplicados en base al nombre del héroe
name_duplicates = df_heroes[df_heroes.duplicated(subset=['nombre'], keep=False)]

# Mostrar el número de duplicados exactos y los duplicados por nombre
exact_duplicates_count = exact_duplicates.shape[0]
name_duplicates_count = name_duplicates.shape[0]

exact_duplicates_count, name_duplicates_count, name_duplicates[['nombre_completo', 'nombre', 'universo']]
```

```python
# Eliminar duplicados en base al nombre del héroe
data_mcu_dc_unique = df_heroes.drop_duplicates(subset=['nombre'], keep='first')

# Verificar el número de héroes únicos después de eliminar duplicados
unique_heroes_corrected = data_mcu_dc_unique['nombre'].nunique()

# Realizar la combinación interna con el dataframe sin duplicados
# Merge attributes for Heroe 1
battle_results_merged_corrected = df_new_results.merge(data_mcu_dc_unique, left_on='Heroe 1', right_on='nombre', how='inner')

# Rename columns to specify Heroe 1 attributes
battle_results_merged_corrected.rename(columns={
    'universo': 'Universe_1',
    'Intelligence': 'Intelligence_1',
    'Strength': 'Strength_1',
    'Speed': 'Speed_1',
    'Durability': 'Durability_1',
    'Power': 'Power_1',
    'Combat': 'Combat_1',
    'superpoderes': 'Superpowers_1',
    'Tier': 'Tier_1',
    'Tier2': 'Tier2_1'
}, inplace=True)

# Drop the extra 'nombre' column
battle_results_merged_corrected.drop(columns=['nombre', 'nombre_completo', 'url', 'heroe'], inplace=True)

# Merge attributes for Heroe 2
battle_results_merged_corrected = battle_results_merged_corrected.merge(data_mcu_dc_unique, left_on='Heroe 2', right_on='nombre', how='inner')

# Rename columns to specify Heroe 2 attributes
battle_results_merged_corrected.rename(columns={
    'universo': 'Universe_2',
    'Intelligence': 'Intelligence_2',
    'Strength': 'Strength_2',
    'Speed': 'Speed_2',
    'Durability': 'Durability_2',
    'Power': 'Power_2',
    'Combat': 'Combat_2',
    'superpoderes': 'Superpowers_2',
    'Tier': 'Tier_2',
    'Tier2': 'Tier2_2'
}, inplace=True)

# Drop the extra 'nombre' column
battle_results_merged_corrected.drop(columns=['nombre', 'nombre_completo', 'url', 'heroe'], inplace=True)
```
